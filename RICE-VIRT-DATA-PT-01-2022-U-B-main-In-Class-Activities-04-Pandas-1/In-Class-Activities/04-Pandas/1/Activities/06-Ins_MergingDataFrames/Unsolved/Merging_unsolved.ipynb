```python
# Dependencies
import pandas as pd
```

```python
# Read in the following data into a DataFrame.
raw_data_info = {
    "customer_id": [112, 403, 999, 543, 123],
    "name": ["John", "Kelly", "Sam", "April", "Bobbo"],
    "email": ["jman@gmail", "kelly@aol.com", "sports@school.edu", "April@yahoo.com", "HeyImBobbo@msn.com"]}

data_info_df=pd.DataFrame(raw_data_info)
data_info_df
```

```python
# Read in the following data into a DataFrame.
raw_data_items = {
    "customer_id": [403, 112, 543, 999, 654],
    "item": ["soda", "chips", "TV", "Laptop", "Cooler"],
    "cost": [3.00, 4.50, 600, 900, 150]}
data_items_df=pd.DataFrame(raw_data_items)
data_items_df
```

```python
# Merge the two dataframes using an inner join (default)
merge_df = pd.merge(data_info_df, data_items_df, on ="customer_id")
merge_df
```

```python
# Merge two dataframes using an outer join
merge_df = pd.merge(data_info_df, data_items_df, on ="customer_id", how = 'outer')
merge_df
```

```python
# Merge two dataframes using a left join
merge_df = pd.merge(data_info_df, data_items_df, on ="customer_id", how = 'left')
merge_df
```

```python
# Merge two dataframes using a right join
merge_df = pd.merge(data_info_df, data_items_df, on ="customer_id", how = 'right')
merge_df
```
