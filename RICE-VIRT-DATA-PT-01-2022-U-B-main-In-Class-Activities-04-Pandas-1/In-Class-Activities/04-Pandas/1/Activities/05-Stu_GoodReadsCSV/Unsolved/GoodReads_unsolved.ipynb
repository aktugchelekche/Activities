```python
# Import Dependencies
import pandas as pd
```

```python
# Make a reference to the books.csv file path
file_load = "Resources/books.csv"

# Import the books.csv file as a DataFrame with utf-8 encoding.

book_df=pd.read_csv(file_load,encoding="ISO-8859-1" )
book_df=pd.DataFrame(book_df)
book_df.head(10)
```

```python
#  Get the data types of each column.
book_df.dtypes
```

```python
# Get a list of all columns within the DataFrame
book_df.columns
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
book_df=book_df[["isbn", "original_publication_year", "original_title", "authors","ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"]]
# Only keep: "isbn", "original_publication_year", "original_title", "authors",
# "ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"
book_df
```

```python
# Rename the headers to be more explanatory
renamed_df = book_df.rename(columns={"isbn": "ISBN",
                                        "original_title": "Original Title",
                                        "original_publication_year": "Publication Year",
                                        "authors": "Authors",
                                        "ratings_1": "One Star Reviews",
                                        "ratings_2": "Two Star Reviews",
                                        "ratings_3": "Three Star Reviews",
                                        "ratings_4": "Four Star Reviews",
                                        "ratings_5": "Five Star Reviews", })
renamed_df.head()
```
